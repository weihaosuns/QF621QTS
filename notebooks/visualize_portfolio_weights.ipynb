```python
import pickle
import os
from config import PORTFOLIO_WEIGHTS_DIR

weights_path = os.path.join(PORTFOLIO_WEIGHTS_DIR, "portfolio_weights_annual_1yr.pkl")
# Load the pickle file (update path accordingly)
with open(weights_path, 'rb') as f:
    all_weights = pickle.load(f)

# all_weights is a dict: {date: DataFrame}
print(f"Total dates loaded: {len(all_weights)}")

# Extract the first date's PCA weights DataFrame
first_date = next(iter(all_weights))
print(f"First rebalance date: {first_date}")

weights_df = all_weights[first_date]

# Show first row of this DataFrame (PC1 weights)
print(weights_df.iloc[1])  # or weights_df.loc['PC1']

# If you want to pretty print all weights for PC1
import pprint
pprint.pprint(weights_df.iloc[1].to_dict())
```

```python
import matplotlib.pyplot as plt

weights = weights_df.iloc[1]  # second PC weights (PC2)

# Separate longs and shorts
longs = weights[weights > 0]
shorts = weights[weights < 0]

fig, ax = plt.subplots(figsize=(12,6))

# Plot longs in green
ax.bar(longs.index, longs.values, color='green', label='Long (Positive Weights)')

# Plot shorts in red
ax.bar(shorts.index, shorts.values, color='red', label='Short (Negative Weights)')

ax.axhline(0, color='black', linewidth=0.8)  # zero line
ax.set_ylabel('Weight')
ax.set_title(f'PC2 Dollar-Neutral Portfolio Weights on {first_date.date()}')
ax.set_xticklabels(weights.index, rotation=90)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

num_pcs_to_plot = 10  # or min(10, len(weights_df))

for i in range(num_pcs_to_plot):
    weights = weights_df.iloc[i]  # PC i (0-based)

    longs = weights[weights > 0]
    shorts = weights[weights < 0]

    fig, ax = plt.subplots(figsize=(12, 4))
    ax.bar(longs.index, longs.values, color='green', label='Long (Positive Weights)')
    ax.bar(shorts.index, shorts.values, color='red', label='Short (Negative Weights)')
    ax.axhline(0, color='black', linewidth=0.8)
    ax.set_ylabel('Weight')
    ax.set_title(f'{weights_df.index[i]} Dollar-Neutral Portfolio Weights on {first_date.date()}')
    ax.set_xticklabels(weights.index, rotation=90)
    ax.legend()
    plt.tight_layout()
    plt.show()
```
