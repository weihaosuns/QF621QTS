```python
import os
import pandas as pd

stock_dir = RAW_STOCK_DIR
ticker = "AAPL"
file_path = os.path.join(stock_dir, f"{ticker}.csv")

# Columns after skipping multi-row header, rename properly
cols = ['Date', 'Price', 'Close', 'High', 'Low', 'Open', 'Volume']

# Read CSV skipping first 3 rows (the multi-header rows)
df = pd.read_csv(file_path, skiprows=3, names=cols, parse_dates=['Date'], index_col='Date')

# Now drop 'Price' column (the weird extra header)
df = df.drop(columns=['Price'])

# Extract Close price column (or Adj Close if available)
price_series = df['Close']

print(price_series.head())
```

```python
import os
import pandas as pd

# Define your stock data folder
stock_dir = RAW_STOCK_DIR

# Loop through each CSV and check for 'Adj Close'
for file in os.listdir(stock_dir):
    if file.endswith('.csv'):
        path = os.path.join(stock_dir, file)
        try:
            df = pd.read_csv(path)
            if 'Adj Close' in df.columns:
                print(f"✅ {file}: 'Adj Close' column found")
            else:
                print(f"❌ {file}: 'Adj Close' column NOT found")
        except Exception as e:
            print(f"[ERROR] Failed to read {file}: {e}")
```
